## 使用plot()绘制折线图

```python
import matplotlib.pyplot as plt
squares = [1, 4, 9, 16, 25]
```

```python
print(squares)
```

```python
plt.plot(squares)
```

```python
plt.plot(range(1,6),squares)
```

```python
plt.plot(range(1,6),squares, linewidth=5)
plt.title("Squares Numbers",fontsize=24)
plt.xlabel("Value",fontsize=14)
plt.ylabel("Square of Value",fontsize=14)
plt.tick_params(axis="both",labelsize=14)
```

## 使用scatter()绘制散点图

```python
plt.scatter(2,4,s=2000)
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value",fontsize=14)
plt.tick_params(axis="both",labelsize=14)
plt.show()
```

```python
x_value=[1,2,3,4,5]
y_value=[1,4,9,16,25]

plt.scatter(x_value,y_value)
```

```python
x_value=list(range(1,1001))
y_value=[x**2 for x in x_value]

plt.scatter(x_value,y_value, c="green", edgecolors="none",s=10)

plt.axis([0,1100,0,1100000])
```

```python
x_value=list(range(1,1001))
y_value=[x**2 for x in x_value]

plt.scatter(x_value,y_value, c="skyblue", edgecolors="none",s=10)

plt.axis([0,1100,0,1100000])
```

### 使用颜色映射

**我们将参数c 设置成了一个 y 值列表， 并使用参数cmap 告诉pyplot 使用哪个颜色映射。 这些代码将 y 值较小的点显示为浅蓝色， 并将 y 值较大的点显示为深蓝色**

```python
x_value=list(range(1,1001))
y_value=[x**2 for x in x_value]

plt.scatter(x_value,y_value, c=y_value, cmap=plt.cm.Blues,edgecolors="none",s=10)

plt.axis([0,1100,0,1100000])
```
